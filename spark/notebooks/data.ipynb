```python
df = pd.read_csv("en.openfoodfacts.org.products.csv", sep='\t', nrows=1000)
```

```python
df = df.dropna(thresh=300, axis = 1)
```

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
# Определяем, какой тип данных вы хотите удалить
dtype_to_drop = 'object'

# Получаем список столбцов с нужным типом данных
columns_to_drop = df.select_dtypes(include=[dtype_to_drop]).columns

# Удаляем столбцы с указанным типом данных
df = df.drop(columns=columns_to_drop)
```
